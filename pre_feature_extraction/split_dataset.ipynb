```python
import json
import pandas as pd
import ast
```

```python
df = pd.read_csv("../processed_datasets/chanlamcoupling_canonicalsmiles.csv")

df.head(5)
```

```python
rf_ratio = 0.6
val_ratio = 0.2
mt_ratio = 0.2


train_df = df.sample(frac=rf_ratio, random_state=42)
remaining_df = df.drop(train_df.index)

val_df = remaining_df.sample(frac=val_ratio / (val_ratio + mt_ratio), random_state=42)

mt_df = remaining_df.drop(val_df.index)


train_df.head(2)
val_df.head(2)
mt_df.head(2)


train_df.to_csv("../dataset_splits/rf_train_barebones.csv", index=False)
val_df.to_csv("../dataset_splits/rf_penultimate_eval.csv", index=False)
mt_df.to_csv("../dataset_splits/mt_train.csv", index=False)
```
