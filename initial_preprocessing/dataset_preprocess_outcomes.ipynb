```python
import json
import pandas as pd
```

```python
df = pd.read_csv("raw_datasets/chanlamcoupling_all_flat_tidied_preprocessedexcludingoutcomes.csv")
df
```

```python
import ast
```

```python
results = []

data = df["outcomes"]

print(data)

product_1_smiles = []
product_1_yield = []
product_2_smiles = []
product_2_yield = []

for i in range(0, len(data)):
    try:
        temp = ast.literal_eval(data[i])
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing JSON data at index {i}: {e}")
        break
    
    
    
    reactions = temp[0]["products"]
    reaction_1 = reactions[0]
    reaction_2 = reactions[1]

    product_1 = reaction_1["identifiers"][0]["value"]
    product_2 = reaction_2["identifiers"][0]["value"]

    # print(product_1)
    # print(product_2)

    yield_1 = reaction_1["measurements"][0]["percentage"]["value"]
    yield_2 = reaction_2["measurements"][0]["percentage"]["value"]

    # print(product_1, yield_1)
    # print(product_2, yield_2)

    product_1_smiles.append(product_1)
    product_1_yield.append(yield_1)
    product_2_smiles.append(product_2)
    product_2_yield.append(yield_2)


   

df["product_1_smiles"] = product_1_smiles
df["product_1_yield"] = product_1_yield
df["product_2_smiles"] = product_2_smiles
df["product_2_yield"] = product_2_yield

df.head(5)

```

```python
df.drop(columns=["boronic_acid_ratio", "sulfonamide_ratio", "base_ratio", "catalyst_ratio"], inplace=True)
df.to_csv("processed_datasets/chanlamcoupling_all_flat_tidied_preprocessedincludingoutcomes.csv", index=False)
```
