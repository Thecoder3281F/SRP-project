```python
import json
import pandas as pd
import ast
```

```python
from rdkit import Chem
```

```python
df = pd.read_csv("../processed_datasets/chanlamcoupling_all_flat_tidied_preprocessedincludingoutcomes_cleaned_shuffled.csv")
df.head(5)
```

```python
df.loc[df["base_smiles"].isna(), "base_smiles"] = "NoBase"
df.head(5)
```

```python
solvent_canonical_smiles = []
boronic_acid_canonical_smiles = []
sulfonamide_canonical_smiles = []
base_canonical_smiles = []
catalyst_canonical_smiles = []
product_1_canonical_smiles = []
product_2_canonical_smiles = []

for row in df.itertuples(index=False):
    solvent_smiles = row.solvent_smiles
    boronic_acid_smiles = row.boronic_acid_smiles
    sulfonamide_smiles = row.sulfonamide_smiles
    base_smiles = row.base_smiles
    catalyst_smiles = row.catalyst_smiles
    product_1_smiles = row.product_1_smiles
    product_2_smiles = row.product_2_smiles

    solvent_mol = Chem.MolFromSmiles(solvent_smiles)
    boronic_acid_mol = Chem.MolFromSmiles(boronic_acid_smiles)
    sulfonamide_mol = Chem.MolFromSmiles(sulfonamide_smiles)
    catalyst_mol = Chem.MolFromSmiles(catalyst_smiles)
    product_1_mol = Chem.MolFromSmiles(product_1_smiles)
    product_2_mol = Chem.MolFromSmiles(product_2_smiles)

    if base_smiles != "NoBase":
        base_mol = Chem.MolFromSmiles(base_smiles)
        base_canonical_smiles.append(Chem.MolToSmiles(base_mol, canonical=True))
    else:
        base_canonical_smiles.append("NoBase")


    solvent_canonical_smiles.append(Chem.MolToSmiles(solvent_mol, canonical=True))
    boronic_acid_canonical_smiles.append(Chem.MolToSmiles(boronic_acid_mol, canonical=True))
    sulfonamide_canonical_smiles.append(Chem.MolToSmiles(sulfonamide_mol, canonical=True))
    catalyst_canonical_smiles.append(Chem.MolToSmiles(catalyst_mol, canonical=True))
    product_1_canonical_smiles.append(Chem.MolToSmiles(product_1_mol, canonical=True))
    product_2_canonical_smiles.append(Chem.MolToSmiles(product_2_mol, canonical=True))


df["solvent_canonical_smiles"] = solvent_canonical_smiles
df["boronic_acid_canonical_smiles"] = boronic_acid_canonical_smiles
df["sulfonamide_canonical_smiles"] = sulfonamide_canonical_smiles
df["base_canonical_smiles"] = base_canonical_smiles
df["catalyst_canonical_smiles"] = catalyst_canonical_smiles
df["product_1_canonical_smiles"] = product_1_canonical_smiles
df["product_2_canonical_smiles"] = product_2_canonical_smiles
```

```python
df.drop(columns=["solvent_smiles", "boronic_acid_smiles", "sulfonamide_smiles", "base_smiles", "catalyst_smiles", "product_1_smiles", "product_2_smiles"], inplace=True)
df.head(5)

df.to_csv("../processed_datasets/chanlamcoupling_canonicalsmiles.csv", index=False)
```
